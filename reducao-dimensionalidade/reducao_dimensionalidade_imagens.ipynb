Convertendo Imagens Coloridas para Tons de Cinza

```python
def rgb_to_grayscale(img_path):
  """Converte uma imagem colorida para tons de cinza.

  Args:
    img_path: Caminho para a imagem.

  Returns:
    Uma imagem em tons de cinza.
  """

  # Lê a imagem colorida
  img = cv2.imread(img_path)

  # Converte para tons de cinza
  gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  return gray_img
```

```python
def rgb_to_grayscale_weighted(img_path):
  """Converte uma imagem colorida para tons de cinza usando a média ponderada.

  Args:
    img_path: Caminho para a imagem.

  Returns:
    Uma imagem em tons de cinza.
  """

  # Lê a imagem colorida
  img = cv2.imread(img_path)

  # Define os pesos para cada canal (valores comuns)
  weights = [0.299, 0.587, 0.114]

  # Calcula a média ponderada para cada pixel
  gray_img = np.dot(img[..., :3], weights)

  return gray_img.astype(np.uint8)  # Converte para uint8 para representar valores de 0 a 255
```

Convertendo Imagens Coloridas para Imagem Binarizada (preto e branco)

```python
def rgb_to_binary(img_path, threshold=127):
  """Converte uma imagem colorida para binária usando um limiar.

  Args:
    img_path: Caminho para a imagem.
    threshold: Valor do limiar (0-255).

  Returns:
    Uma imagem binária.
  """

  # Lê a imagem e converte para tons de cinza
  gray_img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

  # Binariza a imagem
  ret, binary_img = cv2.threshold(gray_img, threshold, 255, cv2.THRESH_BINARY)

  return binary_img
```
